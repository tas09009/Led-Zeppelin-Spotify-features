```python
# Import packages

from spotipy.oauth2 import SpotifyClientCredentials
import spotipy

sp = spotipy.Spotify(client_credentials_manager=SpotifyClientCredentials())
```

```python
# Define the two songs to compare
daechwita = '0H4ugk6rhnXmTl47ayy9O5'
dna = '6HZILIRieu8S0iqY8kIKhj'
```

## Song features + analysis

```python
dae_fea = sp.audio_features(daechwita)[0]
dae_fea
```

```python
dae_ana = sp.audio_analysis(daechwita)
dae_ana
```

```python
dna_fea = sp.audio_features(dna)[0]
dna_fea
```

```python
dna_ana = sp.audio_analysis(dna)
dna_ana
```
